```python
import matplotlib.pyplot as plt
```

```python
import torch
from torch import nn
import torch.nn.functional as F
```

```python
from torchvision import datasets
from torchvision.transforms import ToTensor
from torch.utils.data import DataLoader
```

```python
import torchvision.utils as utils
import torchvision.transforms as transforms

standardizator = transforms.Compose([
                    transforms.ToTensor(),
                    transforms.Normalize(mean=(0.5),
                                         std=(0.5))]) 
```

```python
train_data = datasets.MNIST(
    root='data',
    train=True,
    download=True,
    transform=standardizator
)

test_data = datasets.MNIST(
    root='data',
    train=False,
    download=True,
    transform=standardizator
)

batch_size = 200
train_dataloader = DataLoader(train_data, batch_size=batch_size)
test_dataloader = DataLoader(test_data, batch_size=batch_size)
```

```python
device = ('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
d_noise = 100
d_hidden = 256
```

```python
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.generate = nn.Sequential(
            nn.Linear(d_noise, d_hidden),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(d_hidden, d_hidden),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(d_hidden, 28*28),
            nn.Tanh()
        )
    
    def forward(self, x):
        g_x = self.generate(x)
        return g_x
```

```python
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.discriminate = nn.Sequential(
            nn.Linear(28*28, d_hidden),
            nn.LeakyReLU(),
            nn.Dropout(0.1),
            nn.Linear(d_hidden, d_hidden),
            nn.LeakyReLU(),
            nn.Dropout(0.1),
            nn.Linear(d_hidden, 1),
            nn.Sigmoid()
        )
    
    def forward(self, x):
        d_x = self.discriminate(x)
        return d_x
```

```python
def loss_fn_d(d_x, d_g_z):
    return -1 * (torch.log(d_x) + torch.log(1. - d_g_z))
```

```python
def loss_fn_g(d_g_z):
    return -1 * torch.log(d_g_z).mean()
```

```python
def sample_z(batch_size=1, d_noise=100):
    return torch.randn(batch_size, d_noise, device=device)
```

```python
def train(dataloader, G, D, optimizer_g, optimizer_d):
    G.train()
    D.train()
    for batch, (x, _) in enumerate(dataloader):
        x = x.to(device)
        z = sample_z(batch_size, d_noise)
        g_z = G(z)
        d_x = D(x.view(-1, 28*28))
        d_g_z = D(g_z)
        
        loss_d = loss_fn_d(d_x, d_g_z).mean()
        optimizer_d.zero_grad()
        loss_d.backward()
        optimizer_d.step()
        
        d_g_z = D(G(sample_z(batch_size, d_noise)))
        
        loss_g = loss_fn_g(d_g_z)
        optimizer_g.zero_grad()
        loss_g.backward()
        optimizer_g.step()
```

```python
def test(dataloader, G, D):
    num_batches = len(dataloader)
    real, fake = 0, 0
    G.eval()
    D.eval()
    
    with torch.no_grad():
        for batch, (x, _) in enumerate(dataloader):
            x = x.to(device)
            z = sample_z(batch_size, d_noise)
            g_z = G(z)
            d_x = D(x.view(-1, 28*28))
            d_g_z = D(g_z)
            real += d_x.sum().item()
            fake += d_g_z.sum().item()
    
    real /= 10_000
    fake /= 10_000
            
    return real, fake
```

```python
def init_params(model):
    for p in model.parameters():
        if(p.dim() > 1):
            nn.init.xavier_normal_(p)
        else:
            nn.init.uniform_(p, 0.1, 0.2)
```

```python
G = Generator().to(device)
D = Discriminator().to(device)
```

```python
z = sample_z(batch_size)

print(G(z).shape)
print(D(G(z)).shape)
print(D(G(z)[0:5]).transpose(0,1))
```

```python
init_params(G)
init_params(D)

learning_rate = 2e-4
optimizer_g = torch.optim.Adam(G.parameters(), lr=learning_rate)
optimizer_d = torch.optim.Adam(D.parameters(), lr=learning_rate)
```

```python
real_trace = []
fake_trace = []
epochs = 200
for i in range(epochs):
    train(train_dataloader, G, D, optimizer_g, optimizer_d)
    real, fake = test(test_dataloader, G, D)
    real_trace.append(real)
    fake_trace.append(fake)
    
    if ((i+1) % 50) == 0:
        print(f'epoch {i+1}/{epochs} real: {real} fake: {fake}')
        samples = torch.Tensor.cpu(G(sample_z(8)).view(-1, 1, 28, 28)).detach().numpy()
        figure, plot = plt.subplots(1, 8, figsize=(4, 2))
        for i in range(8):
            plot[i].imshow(samples[i][0], cmap='gray')
            plot[i].axis('off')
        plt.show()
```

```python
plt.plot(real_trace, label='D(x)')
plt.plot(fake_trace, label='D(G(z))')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
