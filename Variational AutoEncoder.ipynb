```python
import torch
from torch import nn
import torch.nn.functional as F
```

```python
from torchvision import datasets
from torchvision.transforms import ToTensor
from torch.utils.data import DataLoader
```

```python
train_data = datasets.MNIST(
    root='data',
    train=True,
    download=True,
    transform=ToTensor()
)

test_data = datasets.MNIST(
    root='data',
    train=False,
    download=True,
    transform=ToTensor()
)

train_dataloader = DataLoader(train_data, batch_size=100)
test_dataloader = DataLoader(test_data, batch_size=100)
```

```python
device = ('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
class VAE(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.encode = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 256),
            nn.ReLU(),
        )
        self.mu = nn.Linear(256, 10)
        self.log_var = nn.Linear(256, 10)
        self.decode = nn.Sequential(
            nn.Linear(10, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, 28*28),
            nn.Sigmoid()
        )
    
    def reparameterize(self, mu, log_var):
        std = torch.exp(0.5*log_var)
        eps = torch.randn_like(std)
        return eps.mul(std).add(mu)
    
    def forward(self, x):
        x_ = self.flatten(x)
        x_ = self.encode(x_)
        mu = self.mu(x_)
        log_var = self.log_var(x_)
        z = self.reparameterize(mu, log_var)
        recon_x = self.decode(z).reshape(x.shape)
        return recon_x, mu, log_var
```

```python
def vae_loss(recon_x, x, mu, log_var):
    reconstruction_error = bceloss(recon_x, x)
    regularization = 0.5 * torch.sum(mu.pow(2) + log_var.exp() - log_var - 1)
    return reconstruction_error, regularization
```

```python
def train(dataloader, model, optimizer):
    size = len(dataloader.dataset)
    for batch, (x, _) in enumerate(dataloader):
        x = x.to(device)
        recon_x, mu, log_var = model(x)
        rec, reg = vae_loss(recon_x, x, mu, log_var)
        loss = rec + reg
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if batch % 100 == 0:
            loss, current = loss.item(), (batch+1) * len(x)
            print(f'loss: {loss/len(x):>7f} rec: {rec:>3f} reg: {reg:>3f} [{current:>5d}/{size:>5d}]')
```

```python
def test(dataloader, model):
    num_batches = len(dataloader)
    test_loss = 0
    
    with torch.no_grad():
        for batch, (x, _) in enumerate(dataloader):
            x = x.to(device)
            recon_x, mu, log_var = model(x)
            rec, reg = vae_loss(recon_x, x, mu, log_var)
            loss = rec + reg
            test_loss += loss
    
    if k % 5 == 0:
        f, a = plt.subplots(2, 5, figsize=(5, 2))
        for i in range(5):
            img = torch.Tensor.cpu(x)[i][0]
            a[0][i].imshow(img, cmap='gray')
            a[0][i].set_xticks(()); a[0][i].set_yticks(())
            img2 = torch.Tensor.cpu(recon_x)[i][0]
            a[1][i].imshow(img2, cmap='gray')
            a[1][i].set_xticks(()); a[1][i].set_yticks(())
    
    test_loss /= num_batches
    print(f'Avg loss: {test_loss}')
```

```python
model = VAE().to(device)
learning_rate = 1e-3
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
bceloss = nn.BCELoss(reduction='sum')
```

```python
k = 0
epochs = 50
for i in range(epochs):
    print(f'Epoch {i+1}\n---------------------------')
    train(train_dataloader, model, optimizer)
    test(test_dataloader, model)
    k += 1
```
